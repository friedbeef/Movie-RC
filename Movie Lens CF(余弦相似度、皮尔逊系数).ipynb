# 读取、处理、转换

```python
import pandas as pd
```

```python
ratings=pd.read_csv('C:/Users/78569/Desktop/电影推荐系统/data/ratings.csv')
movies=pd.read_csv('C:/Users/78569/Desktop/电影推荐系统/data/movies.csv')
```

### 合并成一个df
#### merge()

```python
ratings=pd.merge(ratings,movies)
```

```python
ratings.head()
```

### 去掉不用的列
#### drop()

```python
ratings=ratings.drop(['timestamp','genres'],axis=1)
```

```python
ratings.head()
```

### 转换成类似toy的格式
#### pivot_table()

```python
user_ratings=ratings.pivot_table(index=['userId'],columns=['title'],values=['rating'])
user_ratings.head()
```

### 去掉少于10人评分的电影，剩余的缺失值以0填充，dropna()

#### dropna()中的thresh = a，代表非Na值需要大于等于a个，不然就会被drop
#### 即drop掉非Na值小于a的

```python
user_ratings=user_ratings.dropna(thresh=10,axis=1).fillna(0)
```

```python
user_ratings.head()
```
